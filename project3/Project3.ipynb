### Part a), setting up the problem

The physical problem can be that of the temperature gradient in a rod of length $L=1$ at $x=0$ and $x=1$.
We are looking at a one-dimensional
problem

$$
\frac{\partial^2 u(x,t)}{\partial x^2} =\frac{\partial u(x,t)}{\partial t}, t> 0, x\in [0,L]
$$

or

$$
u_{xx} = u_t,
$$

with initial conditions, i.e., the conditions at $t=0$,

$$
u(x,0)= \sin{(\pi x)} \hspace{0.5cm} 0 < x < L,
$$

with $L=1$ the length of the $x$-region of interest. The 
boundary conditions are

$$
u(0,t)= 0 \hspace{0.5cm} t \ge 0,
$$

and

$$
u(L,t)= 0 \hspace{0.5cm} t \ge 0.
$$

The function $u(x,t)$  can be the temperature gradient of a  rod.
As time increases, the velocity approaches a linear variation with $x$. 

We will limit ourselves to the so-called explicit forward Euler algorithm with discretized versions of time given by a forward formula and a centered difference in space resulting in

$$
u_t\approx \frac{u(x,t+\Delta t)-u(x,t)}{\Delta t}=\frac{u(x_i,t_j+\Delta t)-u(x_i,t_j)}{\Delta t}
$$

and

$$
u_{xx}\approx \frac{u(x+\Delta x,t)-2u(x,t)+u(x-\Delta x,t)}{\Delta x^2},
$$

or

$$
u_{xx}\approx \frac{u(x_i+\Delta x,t_j)-2u(x_i,t_j)+u(x_i-\Delta x,t_j)}{\Delta x^2}.
$$

Write down the algorithm and the equations you need to implement.
Find also the analytical solution to the problem.

### Convolution layer

```python
def padding(image, kernel):
    # calculate r and c
    r = (kernel.shape[0] // 2) * 2
    c = (kernel.shape[1] // 2) * 2
    
    # padded image dimensions
    padded_height = image.shape[0] + r
    padded_width = image.shape[1] + c
    
    # for more readable code
    k_half_height = kernel.shape[0] // 2
    k_half_width = kernel.shape[1] // 2

    # zero matrix with padded dimensions
    padded_img = np.zeros((padded_height, padded_width))

    # place image into zero matrix
    padded_img[k_half_height : padded_height - k_half_height,
               k_half_width : padded_width - k_half_width] = image[:, :]

    return padded_img
    
def convolve(image, kernel, stride=1):
    for i in range(2):
        kernel = np.rot90(kernel)

    k_half_height = kernel.shape[0] // 2
    k_half_width = kernel.shape[0] // 2

    conv_image = np.zeros(image.shape)
    pad_image = padding(image, kernel)

    for i in range(k_half_height, conv_image.shape[0] + k_half_height, stride):
        for j in range(k_half_width, conv_image.shape[1] + k_half_width, stride):
            conv_image[i - k_half_height, j - k_half_width] = np.sum(
                pad_image[
                    i - k_half_height : i + k_half_height + 1, j - k_half_width : j + k_half_width + 1
                ]
                * kernel
            )

    return conv_image

def generate_gauss_mask(sigma, K=1):
    side = np.ceil(1 + 8 * sigma)
    y, x = np.mgrid[-side // 2 + 1 : (side // 2) + 1, -side // 2 + 1 : (side // 2) + 1]
    ker_coef = K / (2 * np.pi * sigma**2)
    g = np.exp(-((x**2 + y**2) / (2.0 * sigma**2)))

    return g, ker_coef
```

```python
# test the convolution using the first image in CIFAR dataset
image = train_images[0]
image = image[:, :, 0]  # get the first channel
# show the image
plt.figure(figsize=(5, 5))
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original Image\n(32x32)")
plt.axis("off")
print(f"Shape of the image: {image.shape}")

# define a kernel
g, ker_coef = generate_gauss_mask(1)
gauss_kernel = g * ker_coef

# apply the gauss convolution
conv_image = convolve(image, gauss_kernel)
# show the convolved image
plt.subplot(1, 3, 2)
plt.imshow(conv_image)
plt.title("Gauss Filter\n(32x32)")
plt.axis("off")
print(f"Shape of the convolved (gauss - blurring) image: {conv_image.shape}")

# apply sobel filter
sobel_kernel = np.array([[1, 2, 1],
                    [0, 0, 0], 
                    [-1, -2, -1]])
conv_image = convolve(image, sobel_kernel)
# show the convolved image
plt.subplot(1, 3, 3)
plt.imshow(conv_image)
plt.title("Sobel Filter\n(32x32)")
plt.axis("off")
print(f"Shape of the convolved (sobel - edge detection) image: {conv_image.shape}")

plt.tight_layout()
plt.show()
```

## ROC AUC Curve of all model

```python
from sklearn.metrics import roc_curve, auc
from sklearn.svm import SVC
from sklearn.preprocessing import label_binarize
from sklearn.linear_model import LogisticRegression

# create dataset for this experiment
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0
flattened_test_labels = test_labels.flatten()
test_labels = tf.keras.utils.to_categorical(test_labels, 10)

# Assuming test_labels and test_predictions are in shape (n_samples, n_classes)
n_classes = test_labels.shape[1]
print(f"Number of classes: {n_classes}")

# init the plot
plt.figure()

def plot_roc(test_predictions, model_name, test_labels=test_labels, n_classes=n_classes, line_color='navy'):
    # Initialize dictionaries to hold False Positive Rates (fpr) and True Positive Rates (tpr) for each class
    fpr = {}
    tpr = {}
    roc_auc = {}

    # Calculate fpr, tpr, and AUC for each class
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(test_labels[:, i], test_predictions[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    fpr_grid = np.linspace(0.0, 1.0, 1000)
    # Interpolate all ROC curves at these points
    mean_tpr = np.zeros_like(fpr_grid)
    for i in range(n_classes):
        mean_tpr += np.interp(fpr_grid, fpr[i], tpr[i])  # linear interpolation

    # Average it and compute AUC
    mean_tpr /= n_classes

    # Compute AUC
    macro_auc = auc(fpr_grid, mean_tpr)

    # Plotting the macro-average ROC curve
    plt.plot(fpr_grid, mean_tpr, color=line_color, linestyle='-', linewidth=2,
            label=f'{model_name} (AUC = {macro_auc:.2f})')

############# SVM #############
# create svm model
svm = SVC(kernel='rbf', C=10, gamma=0.001)

# flatten the images
flattened_test_images = test_images.reshape(test_images.shape[0], -1)

# fit the model
svm.fit(flattened_test_images, flattened_test_labels)

# predict the test labels
test_predictions = svm.predict(flattened_test_images)
test_predictions = label_binarize(test_predictions, classes=np.unique(flattened_test_labels))

# plot the ROC curve
plot_roc(test_predictions, 'SVM', line_color='lemonchiffon')

############# Logistic Regression #############
# create logistic regression model
logreg = LogisticRegression(max_iter=2000, solver='liblinear', C=0.032, penalty='l2')

# fit the model
logreg.fit(flattened_test_images, flattened_test_labels)

# predict the test labels
test_predictions = logreg.predict(flattened_test_images)
test_predictions = label_binarize(test_predictions, classes=np.unique(flattened_test_labels))

# plot the ROC curve
plot_roc(test_predictions, 'Logistic Regression', line_color='lightcoral')

############# FFNN ############# 
# create ffnn model
ffnn_model = tf.keras.models.load_model('ffnn_model.keras')

# predict the test labels
test_predictions = ffnn_model.predict(test_images)

# plot the ROC curve
plot_roc(test_predictions, 'FFNN')

############# CNN #############
# create cnn model
cnn_model = tf.keras.models.load_model('vggcnn_model.keras')

# predict the test labels
test_predictions = cnn_model.predict(test_images)

# plot the ROC curve
plot_roc(test_predictions, 'CNN', line_color='aqua')

############# RNN #############
# create rnn model
rnn_model = tf.keras.models.load_model('rnn_model.keras')

# transform the test images
rnn_test_images = test_images.reshape(test_images.shape[0], -1, 3)

# predict the test labels
test_predictions = rnn_model.predict(rnn_test_images)

# plot the ROC curve
plot_roc(test_predictions, 'RNN', line_color='cornflowerblue')

############# CNN RNN #############
# create cnn rnn model
cnn_rnn_model = tf.keras.models.load_model('cnn_rnn_model.keras')

# predict the test labels
test_predictions = cnn_rnn_model.predict(test_images)

# plot the ROC curve
plot_roc(test_predictions, 'CNN RNN', line_color='darkorange')

# Plot the chance level (45-degree line)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Chance level (AUC = 0.5)')
plt.legend(loc='lower right')
plt.title('Macro-average ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```
